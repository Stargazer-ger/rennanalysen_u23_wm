# Analyse der U23-WM Race-Data

```python
# Imports
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt#
import pandas as pd
```

## BW8+

```python
# CSV Heat 1 aufrufen

heat1 = pd.read_csv('race_pdf/2023_u23_wm/bw8+/csv_output/bw8+_heat1.csv')

# CSV Heat 2 aufrufen

heat2 = pd.read_csv('race_pdf/2023_u23_wm/bw8+/csv_output/bw8+_heat2.csv')

# CSV Rep aufrufen

rep = pd.read_csv('race_pdf/2023_u23_wm/bw8+/csv_output/bw8+_rep1.csv')

# CSV Final aufrufen

final = pd.read_csv('race_pdf/2023_u23_wm/bw8+/csv_output/bw8+_final.csv')
```

```python
# Heat 1

# Kopien erstellen und Renndaten zu float konvertieren

heat1_copy = heat1.copy()

race_data = heat1.iloc[3:43, 1:9].astype(float)

# Einteilen der Spalte pro Boot
dist = heat1_copy.iloc[3:43, 0]

heat1_gbr = race_data.iloc[:, 0:2]

heat1_rou = race_data.iloc[:, 2:4]

heat1_ned = race_data.iloc[:, 4:6]

heat1_can = race_data.iloc[:, 6:8]

print(dist.dtypes)
print(heat1_gbr.dtypes)
print(heat1_rou.dtypes)
print(heat1_ned.dtypes)
print(heat1_can.dtypes)
```

```python
# Heat 2

# Kopien erstellen und Renndaten zu float konvertieren

heat2_copy = heat2.copy()

race_data = heat2.iloc[3:43, 1:7].astype(float)

# Einteilen der Spalte pro Boot
dist = heat2_copy.iloc[3:43, 0]

heat2_usa = race_data.iloc[:, 4:6]

heat2_ger = race_data.iloc[:, 2:4]

heat2_ita = race_data.iloc[:, 0:2]



print(dist.dtypes)
print(heat2_usa.dtypes)
print(heat2_ger.dtypes)
print(heat2_ita.dtypes)
```

```python
# Repecharge

# Kopien erstellen und Renndaten zu float konvertieren

rep_copy = rep.copy()

race_data = rep.iloc[3:43, 1:11].astype(float)

# Einteilen der Spalte pro Boot
dist = rep_copy.iloc[3:43, 0]

rep_rou = race_data.iloc[:, 0:2]

rep_can = race_data.iloc[:, 2:4]

rep_ned = race_data.iloc[:, 4:6]

rep_ger = race_data.iloc[:, 6:8]

rep_ita = race_data.iloc[:, 8:10]

print(dist.dtypes)
print(rep_rou.dtypes)
print(rep_can.dtypes)
print(rep_ned.dtypes)
print(rep_ger.dtypes)
print(rep_ita.dtypes)
```

```python
# Final

# Kopien erstellen und Renndaten zu float konvertieren

final_copy = final.copy()

race_data = final.iloc[3:43, 1:].astype(float)

# Einteilen der Spalte pro Boot
dist = final_copy.iloc[3:43, 0]

final_can = race_data.iloc[:, 0:2]

final_ger = race_data.iloc[:, 2:4]

final_usa = race_data.iloc[:, 4:6]

final_gbr = race_data.iloc[:, 6:8]

final_ned = race_data.iloc[:, 8:10]

final_ita = race_data.iloc[:, 10:12]

print(dist.dtypes)
print(final_can.dtypes)
print(final_ger.dtypes)
print(final_usa.dtypes)
print(final_gbr.dtypes)
print(final_ned.dtypes)
print(final_ita.dtypes)
```
